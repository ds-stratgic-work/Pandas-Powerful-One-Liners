# 1.Introduction
One-liners enhance code readability and simplicity, making it easier for users to comprehend and write. These concise yet effective programs can accomplish tasks in just a single line of code. Not all programming languages support one-liners, which is why Python stands out. 


Pandas, a widely-used open-source library in Python, is essential for data scientists dealing with extensive datasets. Its diverse functions, ranging from data reading to cleaning, manipulation, and visualization, simplify data analysis tasks significantly. ely.

```python
# Create DataFrame

import pandas as pd
import numpy as np
import random

import matplotlib.pyplot as plt
import seaborn as sns

city_list = ["Kolkata", "Hyderabad", "Chennai", "Delhi", "Mumbai", 
             "Bangalore", "London", "Moscow", "Dubai", "Tokyo"]

job_list = ["Software Development Engineer", "Research Engineer", 
            "Test Engineer", "Software Development Engineer-II", 
            "Python Developer", "Back End Developer", 
            "Front End Developer", "Data Scientist", 
            "IOS Developer", "Android Developer"]

cmp_list = ["Amazon", "Google", "Infosys", "Mastercard", "Microsoft", 
            "Uber", "IBM", "Apple", "Wipro", "Cognizant"]

df = []
for i in range(1000):
  
    company = random.choice(cmp_list)
    job = random.choice(job_list)
    city = random.choice(city_list)
    salary = int(round(np.random.rand(), 3)*10**6)
    employment = random.choices(["Full Time", "Intern"], weights=(80, 20))[0]
    rating = round((np.random.rand()*5), 1)
    
    df.append([company, job, city, salary, employment, rating])
    
df = pd.DataFrame(df, columns=["Company Name", "Employee Job Title",
                                   "Employee Work Location",  "Employee Salary", 
                                   "Employment Status", "Employee Rating"])

```

# 3. Clean Data
Data cleansing is an essential stage in the data analysis procedure. Pandas provides a variety of functions to facilitate data cleansing, such as eliminating duplicates, handling missing values, and modifying column names.

Below are a few concise commands that can assist you in cleaning your data:

- To remove duplicates: df.drop_duplicates()
- To handle missing values: df.dropna()

```python
# Groupby average
df.groupby('Employee Salary')['Employee Rating'].mean()
```

# 4. Data Manipulation
### Finding the highest Or lowest and all n values in a series
If we are interested in identifying the highest in this dataset, we can achieve this by utilizing the nlargest() function in Pandas. By applying this function, we can obtain the first n rows with the largest values in the specified column(s), arranged in descending order.

It is important to note that nlargest() returns the complete DataFrame, including the columns that were not specified for ordering. Nevertheless, these additional columns do not contribute to the ordering of the DataFrame. The provided code snippet demonstrates the application of the nlargest() function on our DataFrame.

```python
# Largest n values
n = 5

df.nlargest(n, "Employee Salary", keep = "all")
df.head(5)
```

```python
n = 7
df.nsmallest(n, "Employee Rating", keep = "all")
df.head(5)
```

# 5. Cross tabulation
Also known as CrossTabs, enables the computation of the frequency of each combination between two or more columns or series. It organizes the unique values of one column or list as indexes and the unique values of another column or list as column headers. The values within the cells are calculated using an aggregation function, typically representing the co-occurrence frequency.

For example, if we want to determine the count of employees working at each location within each company, we can utilize Crosstab.

```python
pd.crosstab(df["Company"], df["Employee Work Location"])
df.head(5)
```

```python
result_crosstab = pd.crosstab(df["Company"], df["Employee Work Location"])
sns.heatmap(result_crosstab, annot=True)
```

```python
result_crosstab = pd.crosstab(index = df["Company"], 
                              columns=df["Employment Status"], 
                              values = df["Employee Salary"], 
                              aggfunc=np.mean)
                              
sns.heatmap(result_crosstab, annot = True, fmt = 'g')
```

# 6. Pivot tables
Pivot tables serve as a frequently utilized data analysis tool in Excel. In a similar vein to the aforementioned crosstabs, pivot tables in Pandas offer a means to cross-tabulate your data.

While they possess many similarities and are conceptually identical within the realm of Pandas, there exist several differences in implementation that set them apart (for additional information, refer to this source). The provided code snippet exemplifies the utilization of the pivot_table() method to determine the frequency of co-occurrence between the "Company" and "Location".

```python
pd.pivot_table(df, 
               index=["Company"], 
               columns=["Employee Work Location"], 
               aggfunc='size', 
               fill_value=0)
```

# 5. Visualize Data
